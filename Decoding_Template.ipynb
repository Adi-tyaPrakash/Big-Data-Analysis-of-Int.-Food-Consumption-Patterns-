```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df1 = pd.read_csv('nutrient_origin_data_mean_2017-2021_tons_level2_categories.csv')
```

```python
df1.head(10)
```

```python
print(df1.shape)
```

```python
df2 = pd.read_csv('feed_and_food_categories.csv')
df2
```

```python
df3 = pd.read_csv('area_code.csv')
df3
```

```python
convert_to_decimal = lambda x: '{:.9f}'.format(x)
```

```python
exclude_columns = ['Category.Code', 'Primary.Item.Code', 'Area.Code', 'Producer.Country']

for column in df1.columns:
    if column not in exclude_columns:
        df1[column] = df1[column].apply(convert_to_decimal)
```

```python
df1
```

```python
importer_country=[]
for i in df1['Area.Code']:
    for j in df3["Country Area Code"]:
        if i==j:
            Index = df3.loc[df3['Country Area Code'] == j].index[0]
            Value = df3['Country'].loc[Index]
            importer_country.append(Value)
            break
```

```python
exporter_country=[]
for i in df1['Producer.Country']:
    if i==0:
        exporter_country.append('MARS')
        continue
    for j in df3["Country Area Code"]:
        if i==j:
            Index = df3.loc[df3['Country Area Code'] == j].index[0]
            Value = df3['Country'].loc[Index]
            exporter_country.append(Value)
            break
        
```

```python
item_list=[]
for i in df1['Category.Code']:
    for j in df2["Category.code"]:
        if i==j:
            Index = df2.loc[df2['Category.code'] == j].index[0]
            Value = df2['Category.Name'].loc[Index]
            item_list.append(Value)
            break
```

```python
df1["Importer_Country"] = importer_country
df1['Exporter_Country'] = exporter_country
df1['Item_List'] = item_list
df1
```

```python
custom_string = "nan"
for column in df1.columns:
    if column not in exclude_columns:
        df1 = df1[~df1[column].str.contains(custom_string, na=False)]
```

```python
df1
```

```python
df1 = df1[df1['Exporter_Country'] != 'MARS']
```

```python
df1.shape
```

```python
df1 = df1.drop(['Primary.Item.Code'],axis=1)
df1 = df1.drop(['Producer.Country'],axis=1)
df1 = df1.drop(['Area.Code'],axis=1)
df1 = df1.drop(['Category.Code'],axis=1)
```

```python
df1.to_csv("Cleaned_Data.csv",index = False)
```
